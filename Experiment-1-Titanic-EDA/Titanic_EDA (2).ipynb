Part C: Data Dissimilarity for Two Nominal Attributes

```python
df[['Sex', 'Embarked']].head()
```

```python
p1 = df.loc[0, ['Sex', 'Embarked']]
p2 = df.loc[1, ['Sex', 'Embarked']]

matches = (p1 == p2).sum()
total = len(p1)
dissimilarity = 1 - (matches / total)

print("Passenger 1:")
print(p1)

print("\nPassenger 2:")
print(p2)

print("\nDissimilarity =", dissimilarity)
```

Part D: Data Dissimilarity for Two Numeric Attributes

```python
df[['Age', 'Fare']].head()
```

```python
import math

p1 = df.loc[0, ['Age', 'Fare']]
p2 = df.loc[1, ['Age', 'Fare']]

distance = math.sqrt(((p1 - p2) ** 2).sum())

print("Passenger 1:")
print(p1)

print("\nPassenger 2:")
print(p2)

print("\nEuclidean Distance =", distance)
```
