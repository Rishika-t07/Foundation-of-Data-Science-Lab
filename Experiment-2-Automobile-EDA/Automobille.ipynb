```python
df = pd.read_csv("Automobile_data (1).csv")
print(df.head())
```

```python
print(df.columns)
```

```python
df["price"] = pd.to_numeric(df["price"], errors="coerce")
df["horsepower"] = pd.to_numeric(df["horsepower"], errors="coerce")
df["bore"] = pd.to_numeric(df["bore"], errors="coerce")
df["stroke"] = pd.to_numeric(df["stroke"], errors="coerce")
df["peak-rpm"] = pd.to_numeric(df["peak-rpm"], errors="coerce")
df["normalized-losses"] = pd.to_numeric(df["normalized-losses"], errors="coerce")
```

```python
print(df.info())
```

```python
plt.hist(df["price"].dropna(), bins=10)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
make = df["make"].value_counts()

plt.bar(make.index, make.values)
plt.xticks(rotation=90)
plt.title("Number of Cars by Make")
plt.xlabel("Car Make")
plt.ylabel("Count")
plt.show()
```

```python
plt.scatter(df["engine-size"], df["price"])
plt.title("Engine Size vs Price")
plt.xlabel("Engine Size")
plt.ylabel("Price")
plt.show()
```

```python
plt.boxplot(df["city-mpg"])
plt.title("City MPG")
plt.ylabel("Mileage")
plt.show()
```

```python
fuel = df["fuel-type"].value_counts()

plt.pie(fuel.values, labels=fuel.index, autopct="%1.1f%%")
plt.title("Fuel Type Distribution")
plt.show()
```

```python
df = df.sort_values("highway-mpg")

plt.plot(df["highway-mpg"], df["price"])
plt.title("Highway MPG vs Price")
plt.xlabel("Highway MPG")
plt.ylabel("Price")
plt.show()
```

```python
import math

car1 = df.loc[0, ['engine-size', 'price']]
car2 = df.loc[1, ['engine-size', 'price']]

distance = math.sqrt(((car1 - car2) ** 2).sum())

print("Car 1:")
print(car1)

print("\nCar 2:")
print(car2)

print("\nEuclidean Distance =", distance)
```

```python
import math

car1 = df.loc[0, ['engine-size', 'price']]
car2 = df.loc[1, ['engine-size', 'price']]

distance = math.sqrt(((car1 - car2) ** 2).sum())

similarity = 1 / (1 + distance)

print("Car 1:")
print(car1)

print("\nCar 2:")
print(car2)

print("\nDissimilarity (Euclidean Distance) =", distance)
print("Similarity Score =", similarity)
```
